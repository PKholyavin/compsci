### Задание 1

Напишите программу, которая принимает на вход слова через `input()` до тех пор, пока пользователь не введёт слово "stop". После этого нужно будет в порядке алфавита вывести все первые буквы, которые встретились, и для каждой буквы &ndash; список слов на эту букву. Для этого:

1. Заведите пустой словарь
2. Начните бесконечный цикл, который будет останавливаться по слову "stop"
3. У каждого слова возьмите первую букву и запишите её в отдельную переменную
4. Проверьте, есть ли эта буква в словаре (с помощью оператора вхождения `in`)
5. Если нет, сделайте её новым ключом, а значением &ndash; пустой список
6. Добавьте слово в список, который лежит в словаре по этой букве

7. После конца цикла заведите другой цикл, который перебирает все буквы в словаре в алфавитном порядке
8. Напечатайте букву в верхнем регистре (`.upper()`)
9. В алфавитном порядке выведите все слова, которые лежат в списке, доступном по этой букве в словаре &ndash; каждое на своей строчке
