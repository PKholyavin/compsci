### Задания для самостоятельного выполнения

#### Задание 1

Напишите программу, которая запрашивает у пользователя цену товара и процент скидки, а потом выводит на экран итоговую стоимость товара.

#### Задание 2

Напишите программу, которая вычисляет индекс массы тела человека. Пусть пользователь введёт с клавиатуры два числа: массу в кг и рост в сантиметрах. ИМТ вычисляется как масса в кг, делённая на рост в метрах в квадрате.

#### Задание 3

Напишите программу, которая вычисляет площадь треугольника по формуле Герона:

$$ S = \sqrt{p \cdot (p - a) \cdot (p - b) \cdot (p - c)} $$

где _p_ &mdash; полупериметр, _a, b, c_ &mdash; стороны (вводятся пользователем с клавиатуры).

Помните, что квадратный корень можно заменить на возведение в степень 0.5.

#### Задание 4

Напишите программу, которая повторяет слово заданное количество раз. Пусть пользователь введёт сначала слово, а потом число. Например, если пользователь вводит "привет" и 3, программа должна написать "приветприветпривет".

#### Задание 5

Напишите программу, которая вычисляет объём конуса по радиусу основания и высоте (примите $\pi=3.14$):

$$ V = \frac{1}{3} \pi r^2 h$$

#### Задание 6

Напишите программу для вычисления n-ного члена и суммы членов арифметической прогресии. Пусть пользователь вводит первый член, шаг и n (номер последнего члена).

$$ a_n = a_1 + d (n - 1) $$

$$ S_n = \frac{a_1 + a_n}{2}n $$
