### Генерация самого вероятного предложения на основе текста

Напишите программу, которая анализирует большой текст (например, книгу) и строит на его основе вероятностную модель слов — так называемую биграммную модель.

Биграммы — это пары слов, идущие друг за другом в тексте. Например, в предложении:

    "Мама мыла раму"

биграммы будут:

    ("Мама", "мыла")

    ("мыла", "раму")

Биграммная модель хранит, какие слова чаще всего следуют за каждым словом. Это позволяет предсказывать следующее слово с учётом предыдущего.

Что должна делать программа:

1. Прочитать текст из файла (.txt).

2. Разбить текст на слова (можно использовать `split()`).

3. Построить словарь биграмм, где:

Ключ — слово

Значение — словарь всех слов, которые шли после него, с количеством вхождений

4. Случайным образом выбрать первое слово.

5. Постепенно строить предложение, каждый раз выбирая наиболее вероятное следующее слово (т.е. то, что встречается чаще всего после текущего). Добавлять слова в список.

6. Завершить построение, если выбранное слово заканчивается на точку, вопросительный или восклицательный знак или длина предложения превышает 15 слов.

7. Соединить список с помощью `join()` и вывести получившееся предложение на экран.

3. Как создать словарь биграмм:

```
Считать текст из файла и сохранить в переменную text
Привести текст к нижнему регистру
Разбить текст на список слов (например, по пробелам) и сохранить в переменную words

Создать пустой словарь bigrams

ДЛЯ КАЖДОГО индекса i от 0 до длины списка words - 1 (не включая):
    текущее_слово = words[i]
    следующее_слово = words[i + 1]

    ЕСЛИ текущее_слово не находится в словаре bigrams:
        добавить текущее_слово в словарь bigrams и присвоить ему пустой словарь

    ЕСЛИ следующее_слово не находится в словаре bigrams[текущее_слово]:
        добавить следующее_слово в словарь bigrams[текущее_слово] со значением 1
    ИНАЧЕ:
        увеличить значение bigrams[текущее_слово][следующее_слово] на 1
```
