### Практические задания на цикл for

#### Задание 1

Пусть пользователь вводит с клавиатуры целое число N. Если оно положительное, найти все его делители и вывести на экран. Иначе вывести сообщение об ошибке.

Идея: перебрать в цикле все числа от 1 до N. Если N делится на это число без остатка, вывести число на экран.

(*) Подумайте, как можно сократить цикл. Обязательно ли перебирать все числа до N?

#### Задание 2

Любое натуральное число является недостаточным (сумма собственных делителей которого меньше его самого), совершенным (сумма собственных делителей равна числу) или избыточным (сумма собственных делителей больше его самого). Собственные делители &mdash; это все делители числа, кроме его самого.

Пусть пользователь вводит с клавиатуры число N. Определите, является ли оно недостаточным, совершенным или избыточным.

Идея: определить в цикле сумму делителей (см. задание 1) и сравнить с самим числом. Заведите до начала цикла переменную, равную 0. Переберите в цикле все числа от 1 до N. Если N делится на это число без остатка, прибавить его к этой переменной.

#### Задание 3

Пусть пользователь вводит с клавиатуры строку. С помощью цикла `for` переберите все символы в этой строке и определите, совпадает ли количество открывающих и закрывающих скобок.

Идея: до цикла заведём специальный счётчик, равный 0. Каждый раз, когда будем встречать символ `(`, будем увеличивать счётчик на 1, когда будем встречать `)` &ndash; уменьшать. Если после конца цикла счётчик равен 0, выведем текст "Количество ( и ) совпадает", если больше 0 &ndash; "( больше, чем )", если меньше 0 &ndash; "( меньше, чем )".

#### Задание 4

Пусть пользователь вводит с клавиатуры число. Определить сумму его цифр.

Идея: завести переменную, в которой будем хранить сумму. С помощью цикла `for` перебрать все символы в строковом представлении числа, каждую цифру переделать в число с помощью `int()` и прибавить к переменной с суммой.

#### Задание 5

Вывести на экран все числа от 1 до 100. Если число делится на 3, вместо него написать Fizz. Если число делится на 5, вместо него написать Buzz. Если число делится и на 3, и на 5, вместо него написать FizzBuzz. Если число не делится ни на 3, ни на 5, вывести само число.

Дополнительно в конце вывести сумму всех чисел, которые программа ни на что не заменила.

`a % 3 == 0` -> Fizz  
`a % 5 == 0` -> Buzz  
`a % 3 == 0 and a % 5 == 0` -> FizzBuzz

### Домашнее задание

Пусть пользователь вводит с клавиатуры число. Определите, чего в нём больше &mdash; чётных цифр или нечётных.

Подсказка: решается аналогично заданию 3 про скобки. Не забудьте каждую цифру переделать в число с помощью `int()` (см. задание 4).
