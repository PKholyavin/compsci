Задайте двухместные предикаты от (x, y), множества истинности которых представляют собой следующие рисунки: сосиска, пакман, песочные часы, машина, домик, кораблик, 10Б, свой рисунок.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos

# Предикат писать здесь
# ----------
# & = И
# | = ИЛИ
# ~ = НЕ
# ** = возведение в степень
# каждое неравенство нужно заключить в круглые скобки!
# вы можете задать несколько вспомогательных предикатов, чтобы было удобнее
# основной предикат должен называться predicate


def P(x, y):
    return (
        x ** 2 + y ** 2 <= 1
    )


def predicate(x, y):
    return (
        P(x, y) & ((y >= x) | (x <= 0))
    )

# ----------

# Сюда можно поместить название
# ----------
plt.title("Рисунок 1")
# ----------

xlim = (-3, 3)
ylim = (-3, 3)
res = 1000

x = np.linspace(*xlim, res)
y = np.linspace(*ylim, res)

plt.xlim(*xlim)
plt.ylim(*ylim)

xx, yy = np.meshgrid(x, y)

image = predicate(xx, yy)

plt.imshow(np.flip(image, axis=0), extent=(*xlim, *ylim), cmap="gray_r")
plt.show()
```
