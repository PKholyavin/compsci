## Вспоминаем Python

### Задание 1

Напишите программу, которая принимает на вход число и выводит на экран "+", если число больше или равно 18, и "-" в противном случае.

Полезности:
* функция `input()` позволяет прочитать текст, введёный пользователем. В круглых скобках можно написать подсказку для пользователя в кавычках
* функция `int()` позволяет сделать из строки целое число
* для проверки условий используйте оператор `if`

### Задание 2

Напишите программу, которая принимает на вход сначала произвольный текст, а потом число, и выводит введённый текст это количество раз.

* для повторения кода N раз используйте цикл `for` и функцию `range()`

### Задание 3

Напишите программу, которая запрашивает у пользователя цену товара и процент скидки, а потом выводит на экран итоговую стоимость товара.

* функция `float()` позволяет сделать из строки вещественное число

### Задание 4

Напишите программу, которая принимает на вход три положительных числа &ndash; длины сторон треугольника. Программа должна классифицировать его как равносторонний (equilateral), равнобедренный (isosceles) или разносторонний (scalene).

### Задание 5

Напишите программу, которая принимает на вход целое число N, ищет все его делители и выводит на экран.

* используйте цикл `for`, чтобы перебрать все числа d от 1 до N
* используйте оператор "остаток от деления" `%`. Если остаток от деления N на d равен 0, то d &ndash; делитель N.

### Задание 6

Напишите программу, которая принимает на вход пять чисел и выводит на экран их среднее значение.

* чтобы хранить несколько чисел сразу, используйте список `[]` и добавляйте туда числа с помощью метода `.append()`

### Задание 7

Измените предыдущую программу так, чтобы можно было вводить произвольное количество чисел.

* используйте бесконечный цикл
* чтобы остановить ввод чисел, пусть пользователь введёт слово `"stop"`
* в этом случае вы должны остановить цикл с помощью оператора `break`

### Задание 8

Измените предыдущую программу так, чтобы числа нужно было вводить в одну строчку, разделяя их пробелами.

* используйте метод `.split()`, чтобы разделить строку на части
* используйте list comprehension, чтобы превратить список строк в список чисел

### Задание 9

Напишите программу, которая принимает с клавиатуры строку и определяет, совпадает ли количество открывающих и закрывающих скобок.

* используйте цикл `for`, чтобы перебрать все символы в строке
* заведите переменную-счётчик, изначально равную нулю
* увеличивайте её на 1, если встречаете открывающую скобку, и уменьшайте на 1, если встречаете закрывающую
* после конца цикла проверьте: если переменная равна 0, то скобок было равное количество; если больше нуля &ndash; было больше открывающих; если меньше &ndash; было больше закрывающих.

### Задание 10

Напишите программу, которая принимает с клавиатуры последовательность слов, разделённую пробелами. Необходимо убрать из неё повторяющиеся слова и вывести на экран результат.

* заведите пустой список
* по очереди добавляйте туда слова из ввода, используя проверку на вхождение (оператор `in`)
* в конце склейте список в строку с помощью метода `.join()`
