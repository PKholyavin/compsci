#### Задание 1

Напишите программу, которая принимает на вход слово на русском языке, и выводит длину самого длинного сочетания согласных в нём.

(*)

Напишите функцию, которая принимает на вход строку, содержащую текст на русском языке, и возвращает список слов из этого текста, отсортированных в порядке возрастания длины самого длинного сочетания согласных в них. Для этого функцию из пункта (а) передайте в параметр `key` функции `sorted()` или метода списков `.sort()`.

#### Задание 2

Напишите программу для решения квадратных уравнений. Пусть пользователь вводит коэффициенты a, b, c. Вычислите дискриминант и найдите корни, если они есть.

#### Задание 3

Пусть пользователь введёт квадрат из целых чисел 3х3, по 3 числа на каждой строке. В строке числа разделяются пробелами. Определите, является ли квадрат магическим (т.е. сумма чисел по всем вертикалям, горизонталям, диагоналям одинакова).

Пример магического квадрата:
```
8 1 6
3 5 7
4 9 2
```

Измените код так, чтобы сначала пользователь вводил размер своего квадрата, а потом все строчки (т.е. чтобы поддерживались квадраты не только 3х3, но и 4х4, 5х5 и т.д.)

#### Задание 4

Напишите программу, которая принимает на вход слово на русском языке и выводит его копию, в которой все гласные буквы стоят в обратном порядке, а согласные остаются на своих местах. Например: "фонетика" -> "фанитеко"

#### Задание 5

Напишите программу, которая принимает на вход два слова на русском языке и определяет, можно ли второе слово составить из букв первого (с учётом повторяющихся!). Например: "электрификация", "клетка" -> можно; "электрификация", "фифа" -> нельзя.


#### Задание 6

Напишите программу, которая принимает на вход текст на русском языке, и выводит на экран тот же текст, где каждое слово должно быть разделено на слоги в соответствии с принципом открытого слога (открытый слог &mdash; слог, который заканчивается на гласный звук; согласно этому принципу все слоги в слове, кроме, возможно, последнего &mdash; открытые).

Пример:

Был тихий серый вечер. Наш автобус номер семь шёл на запад.

->

Был ти-хий се-рый ве-чер. Наш а-вто-бус но-мер семь шёл на за-пад.
